```python
X=data1.iloc[:,[11,12,1,19,22,23,25,0,2,5]]
y1=y
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier=Sequential()
```

```python
classifier.add(Dense(activation='relu',input_dim=10,units=8,kernel_initializer='uniform'))
```

```python
classifier.add(Dense(activation="relu",  units=8, kernel_initializer="uniform")) # this are my input features
```

```python
classifier.add(Dense(activation='relu',units=6,kernel_initializer='uniform')) # this are the my second hidden layer
```

```python
classifier.add(Dense(activation='sigmoid',units=1,kernel_initializer='uniform')) # here i am using the sigmoid function beacuse
# its a classification problem
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 100)
```

```python
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
cm=confusion_matrix(y_pred,y_test)
cm
```

```python
accuracy=accuracy_score(y_pred,y_test)
accuracy
```

```python
report=classification_report(y_pred,y_test)
report
```
